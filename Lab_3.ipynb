# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

```python
import matplotlib.pyplot as plt

def is_even_or_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    is_even_or_odd(n)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Even/Odd Check")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

```python
def count_greater_than_zero(numbers):
    count = 0
    for num in numbers:
        if num > 0:
            count += 1
    return count

n_values = list(range(1, 101))
time_values = []

for n in n_values:

    data = list(range(-n // 2, n // 2))
    count_greater_than_zero(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity (Count Greater Than Zero)")
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    steps = 0  # To simulate time complexity

    while low <= high:
        steps += 1
        mid = (low + high) // 2
        if arr[mid] == target:
            return steps
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return steps

n_values = [2**i for i in range(1, 15)]
time_values = []

for n in n_values:
    data = sorted(list(range(n)))

    steps_taken = binary_search(data, n - 1)
    time_values.append(steps_taken)

plt.figure()
plt.plot(n_values, time_values, marker='o')
plt.xlabel("Input size n")
plt.ylabel("Time (simulated steps)")
plt.title("O(log n) Logarithmic Time Complexity (Binary Search)")
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", ls="-", alpha=0.6)
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
import random
import math

def merge_sort(arr):
    steps = 0
    if len(arr) <= 1:
        return arr, 0

    mid = len(arr) // 2
    left_half, steps_left = merge_sort(arr[:mid])
    right_half, steps_right = merge_sort(arr[mid:])
    steps += steps_left + steps_right

    i = j = k = 0
    while i < len(left_half) and j < len(right_half):
        steps += 1  # For comparison
        if left_half[i] < right_half[j]:
            arr[k] = left_half[i]
            i += 1
        else:
            arr[k] = right_half[j]
            j += 1
        k += 1

    while i < len(left_half):
        steps += 1
        arr[k] = left_half[i]
        i += 1
        k += 1

    while j < len(right_half):
        steps += 1
        arr[k] = right_half[j]
        j += 1
        k += 1
    return arr, steps

n_values = [2**i for i in range(1, 12)]
time_values = []

for n in n_values:
    data = [random.randint(0, n) for _ in range(n)]
    _, steps_taken = merge_sort(data)
    time_values.append(steps_taken)

plt.figure()
plt.plot(n_values, time_values, marker='o')
plt.xlabel("Input size n")
plt.ylabel("Time (simulated steps)")
plt.title("O(n log n) Log-Linear Time Complexity (Merge Sort)")
plt.grid(True, which="both", ls="-", alpha=0.6)
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
import random
import math

def heapify(arr, n, i, steps):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n:
        steps += 1
        if arr[l] > arr[largest]:
            largest = l

    if r < n:
        steps += 1
        if arr[r] > arr[largest]:
            largest = r

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        steps += 1
        steps = heapify(arr, n, largest, steps)
    return steps

def heap_sort(arr):
    n = len(arr)
    steps = 0


    for i in range(n // 2 - 1, -1, -1):
        steps = heapify(arr, n, i, steps)


    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i] # Swap
        steps += 1 # For swap
        steps = heapify(arr, i, 0, steps)
    return steps

n_values = [2**i for i in range(1, 12)]
time_values = []

for n in n_values:
    data = [random.randint(0, n) for _ in range(n)]
    steps_taken = heap_sort(data.copy())
    time_values.append(steps_taken)

plt.figure()
plt.plot(n_values, time_values, marker='o')
plt.xlabel("Input size n")
plt.ylabel("Time (simulated steps)")
plt.title("O(n log n) Log-Linear Time Complexity (Heap Sort)")
plt.grid(True, which="both", ls="-", alpha=0.6)
plt.xscale('log')
plt.yscale('log')
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
import matplotlib.pyplot as plt
import numpy as np
import random

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

n = np.arange(1, 1000)
n_log_n = n * np.log2(n)

plt.plot(n, n_log_n)
plt.xlabel("n")
plt.ylabel("n log n")
plt.title("O(n log n)")
plt.grid(True)
plt.show()

arr = random.sample(range(1, 100), 10)
print(merge_sort(arr))
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.

```python
import matplotlib.pyplot as plt
import numpy as np

# Input size range
n = np.linspace(1, 50, 100)


y_constant = [1 for _ in n]
y_log_n = np.log2(n)
y_n = n # O(n)
y_n_log_n = n * np.log2(n)
y_n_squared = n**2
y_two_pow_n = 2**n


plt.figure(figsize=(10, 7))

plt.plot(n, y_constant, label='O(1) - Constant', linestyle='--')
plt.plot(n, y_log_n, label='O(log n) - Logarithmic')
plt.plot(n, y_n, label='O(n) - Linear')
plt.plot(n, y_n_log_n, label='O(n log n) - Log-Linear')
plt.plot(n, y_n_squared, label='O(n^2) - Polynomial')

n_exp = np.linspace(1, 15, 100)
y_two_pow_n_limited = 2**n_exp
plt.plot(n_exp, y_two_pow_n_limited, label='O(2^n) - Exponential', color='red', linewidth=2)

plt.xlabel("Input Size (n)")
plt.ylabel("Operations / Time")
plt.title("Comparison of Time Complexities")
plt.legend()
plt.grid(True, linestyle=':')
plt.ylim(0, 1000)
plt.xlim(0, 50)
plt.show()
```

```python
import random
import time
import matplotlib.pyplot as plt

def quick_sort(arr):
    steps = 0
    if len(arr) <= 1:
        return arr, steps

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]


    steps += len(arr)

    sorted_left, steps_left = quick_sort(left)
    sorted_right, steps_right = quick_sort(right)

    steps += steps_left + steps_right
    return sorted_left + middle + sorted_right, steps

n_values = [2**i for i in range(1, 12)]
time_values = []

for n in n_values:

    data = [random.randint(0, n) for _ in range(n)]


    start_time = time.time()
    quick_sort(data.copy())
    end_time = time.time()
    time_values.append((end_time - start_time) * 1000)

plt.figure()
plt.plot(n_values, time_values, marker='o')
plt.xlabel("Input size n")
plt.ylabel("Time (ms)")
plt.title("O(n log n) Average Time Complexity (Quick Sort)")
plt.grid(True, which="both", ls="-", alpha=0.6)
plt.xscale('log')
plt.yscale('log')
plt.show()
```
